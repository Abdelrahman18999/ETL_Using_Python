```python
import pandas as pd
```

# Orders

```python
# orders
orders = pd.read_csv("D:\\ITI\\Python\\ITI_Python_for_Data_Management\\Extraction\\Staging_2\\Orders_Transformed2\\orders_final.csv")
orders.head()
```

```python
orders = orders.drop(['Unnamed: 0', 'status_id'], axis=1)
orders.columns
```

# Order Items

```python
# order items
order_items = pd.read_csv("D:\\ITI\\Python\\ITI_Python_for_Data_Management\\Extraction\\Staging_2\\Transformed_OrderItems_RateX\\Transformed_OrderItems_RateX.csv")
order_items.head()
```

```python
order_items = order_items.drop('Unnamed: 0', axis=1)
order_items.columns
```

# products

```python
# products
products = pd.read_csv("D:\\ITI\\Python\\ITI_Python_for_Data_Management\\Extraction\\Staging_2\\products\\products.csv")
products.head()
```

```python
products = products.drop('Unnamed: 0', axis=1)
products.columns
```

# Merge the three datasets together

```python
merged_df = pd.merge(pd.merge(orders, order_items, left_index=True, right_index=True), products, left_index=True, right_index=True)
merged_df.head()
```

```python
merged_df.columns
```

```python
order_id_x  - y
Extraction_Date_x - y
source_x - y
product_id_x - y
list_price_x - y
```

```python
# Rename the columns
merged_df = merged_df.rename(columns={'order_id_x': 'orderID_orders', 
                                      'order_id_y': 'orderID_items', 
                                      'Extraction_Date_x': 'Extraction_Date_orders',
                                      'Extraction_Date_y': 'Extraction_Date_items',
                                      'source_x': 'source_orders',
                                      'source_y': 'source_items',
                                      'product_id_x': 'product_id_orders',
                                      'product_id_y': 'product_id_items',
                                      'list_price_x': 'list_price_orders',
                                      'list_price_y': 'list_price_items'})

merged_df.columns
```
