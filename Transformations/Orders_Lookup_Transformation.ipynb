```python
import pandas as pd
```

```python
orders = pd.read_csv("Transformed_Data/Orders_Transformed.csv")
orders.head()
```

```python
orders = orders.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
orders.columns
```

```python
# Lookup table for order statuses
status_data = {
    'status_id': [1, 2, 3, 4],
    'status_description': ['Pending', 'Processing', 'Rejected', 'Completed']
}
status_lookup = pd.DataFrame(status_data)
```

```python
status_lookup.head()
```

```python
# Merge the orders DataFrame with the status lookup table
merged_orders = pd.merge(orders, status_lookup, left_on='order_status', right_on='status_id', how='left')
```
