```python
import pandas as pd
```

```python
orders = pd.read_csv("D:\\ITI\\Python\\ITI_Python_for_Data_Management\\Extraction\\Staging_1\\orders\\orders.csv")
orders.head()
```

```python
orders.info()
```

```python
# Convert date columns to datetime
orders['required_date'] = pd.to_datetime(orders['required_date'])
orders['shipped_date'] = pd.to_datetime(orders['shipped_date'])

orders.info()
```

```python
# Add 'late_delivery' column
orders['late_delivery'] = orders['shipped_date'] > orders['required_date']

# Set 'late_delivery' to False if delivery_date is equal to required_date
orders.loc[orders['shipped_date'] == orders['required_date'], 'late_delivery'] = False


orders.sample(10)
```

```python
# Add 'latency_days' column
orders['latency_days'] = (orders['shipped_date'] - orders['required_date']).dt.days
orders.loc[orders['latency_days'] <= 0, 'latency_days'] = None
```

```python
orders.sample(10)
```

```python
# Add 'weekend' column
orders['weekend'] = orders['shipped_date'].dt.dayofweek.isin([6, 0])  # 6=Sunday, 0=Monday

orders.sample(10)
```

```python
orders.to_csv("Orders_Transformed.csv")
```
