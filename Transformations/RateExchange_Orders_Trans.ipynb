```python
import pandas as pd
```

```python
# Load the CSV files into DataFrames
orders_df = pd.read_csv('D:\\ITI\\Python\\ITI_Python_for_Data_Management\\Extraction\\Staging\\order_items\\order_items.csv')
orders_df.head()
```

```python
rateexchange_df = pd.read_csv('D:\\ITI\\Python\\ITI_Python_for_Data_Management\\Extraction\\Staging\\rateexchange\\rateexchange.csv')
rateexchange_df
```

```python
# Add a temporary column with a constant value to use for the cross join
orders_df['tmp'] = 1
rateexchange_df['tmp'] = 1
```

```python
# Merge the DataFrames to create a cross join
merged_df = pd.merge(orders_df, rateexchange_df, on='tmp').drop('tmp', axis=1)

# Calculate the 'list_price_egp' column
merged_df['list_price_egp'] = merged_df['list_price'] * merged_df['EGP']
```

```python
merged_df = merged_df.drop(['Unnamed: 0_x', 'Unnamed: 0_y'], axis=1)
```

```python
merged_df = merged_df.rename(columns={'Extraction_Date_x': 'Extraction_Date_OrderItems'})
merged_df = merged_df.rename(columns={'Extraction_Date_y': 'Extraction_Date_RateExchange'})
```

```python
merged_df = merged_df.rename(columns={'source_x': 'source_OrderItems'})
merged_df = merged_df.rename(columns={'source_y': 'source_RateExchange'})
```
